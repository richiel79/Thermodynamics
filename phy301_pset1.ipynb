```python
# Import relevant packages
from matplotlib import pyplot as plt
from scipy.integrate import simps
from scipy.stats import norm
import random
import numpy as np
```

```python
# Generate Data at Random
i = random.randint(0,100)
data = []

if i % 2 == 0:
    index = random.randint(10000,20000)
    for i in range(index):
            data.append(random.gauss())
else: 
    index = random.randint(4000,8000)
    for i in range(index):
            data.append(random.gauss())


```

```python
# Plot results
plt.hist(data, bins = 100, color = 'skyblue', edgecolor = 'black', density = True)


plt.xlabel('Values')
plt.ylabel('Frequency')
plt.title('PHY301 PSET1')
plt.show()
```

```python
plt.hist(data, bins = 100, color = 'skyblue', edgecolor = 'black', density = True)

xmin,xmax = plt.xlim()
mu, std = norm.fit(data)

x = np.linspace(xmin,xmax,100)
p = norm.pdf(x,mu,std)



plt.xlabel('Values')
plt.ylabel('Frequency')
plt.title('PHY301 PSET1')

##Plots the analytic form of guassian
plt.plot(x,p,'k', linewidth=2)
```

```python

from math import pi
from re import A

##Using a Uniform Distribution
N_draws = 10000
N_sums_random = np.random.randint(1, 101, size = 50)
n_bins = 50

## Keep count of the sum for bins
AVG_SUM = np.zeros(N_draws)

## Iterates over the indices of the array of randomly drawn inputs
for i, N in enumerate(N_sums_random):
    draw = np.random.uniform(0, 1, (N_draws, N))

    sums = np.sum(draw, axis=1)
    AVG_SUM += sums/ len(N_sums_random)

plt.hist(AVG_SUM, bins = n_bins, density = True, alpha = 0.7, edgecolor = 'black')
mu = np.mean(AVG_SUM)
sigma = np.std(AVG_SUM)
x = np.linspace(min(AVG_SUM), max(AVG_SUM), 100)

gauss = (1/(sigma * np.sqrt(2 * pi))) * np.exp(-0.5 * ((x - mu)/sigma)**2)
plt.plot(x, gauss, '-', color = 'red', linewidth=1)

plt.title('Uniform to Normal Histogram')
plt.xlabel('Average Sum')
plt.ylabel('Density')

plt.show()
```
