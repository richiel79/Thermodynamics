```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

Question 6

```python
data_x = [i for i in range(30)]
data_y = [math.factorial(i) for i in range(30)]
y = [(2*math.pi*i)**0.5 *(i/math.e)**i for i in range(30)]


plt.semilogy(data_x,data_y, label='n!', linewidth = 2, color='b')
plt.semilogy(data_x,y, label='Approximation', linestyle ='--', marker="o", alpha = 0.6, linewidth = 2, color='r')

plt.xlabel('n')
plt.ylabel('n!')
plt.title("n! vs. Sterling Approximation")
plt.legend()
plt.show()
```

Testing Approximation

```python
error = 0

for i in range(20):
    i = i+1
    factorial = math.factorial(i)
    approx = i*math.log(i)-i
    
    error = approx/factorial
    if i == 3:
         print('Fractional Error at n = 3:', error)
    elif i == 10:
         print('Fractional Error at n = 10:', error)
    elif i == 19:
        print('Fractional Error at n = 20:', error)
```

Log of Number of States

```python
def f(s,N):
    sigma = N*np.log(N)-s*np.log(s)-(N-s)*np.log(N-s)
    return sigma

N = 10000
states = np.arange(1,10000,1)

plt.plot(2*states-N,f(states,N))
plt.ylabel('ln(Number of States)')
plt.title(r'$\sigma$(E), N = 1000')
plt.show()
```
