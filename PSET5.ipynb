```python
import numpy as np 
import scipy.stats
import matplotlib.pyplot as plt 
```

```python
#initiating variables
x0 = 0
v0 = 0
a = 0.01
b = 0.01 

x=[]
v=[]

n = 100000
```

Part C

```python
x.append(x0)
v.append(v0)

for i in range(n):
    v_step = (1-b)*v[i] - a*x[i] + np.random.normal(0,np.sqrt(2*a*b))
    x_step = x[i] + v_step
    x.append(x_step)
    v.append(v_step)

plt.plot(x)
plt.xlabel("n steps")
plt.ylabel("x(n)")
plt.title("Dimensionless Position X over Time")
plt.show()
```

```python
plt.plot(v, color = 'orange')
plt.xlabel("n steps")
plt.ylabel("v(n)")
plt.title("Dimensionless Velocity V over Time")
plt.show()
```

Part D

```python
#Creating histograms for results and comparing to Boltzman

plt.hist(x[10000:], density = True, bins=100, label = "X Distribution")
plt.xlabel("X Dimionsionless")
plt.ylabel("Frequency")
plt.title("Dimensionless X over Time Distribution")
graph = np.linspace(-5,5, 500)
boltz = scipy.stats.norm.pdf(graph, np.mean(x[1000:]),np.std(x[1000:]))
plt.plot(graph,boltz,label="Boltzman Distribution")
plt.legend()
plt.show()
```

```python
plt.hist(v[1000:], density = True, bins = 100, label="Dimensionless V Distribution")
plt.xlabel("V Dimensionless")
plt.ylabel("Frequency")
plt.title("Dimensionless V over Time Distribution")
graph2 = np.linspace(-0.4,0.4, 500)
bltzman = scipy.stats.norm.pdf(graph2,np.mean(v[1000:]),np.std(v[1000:]))
plt.plot(graph2, bltzman, label="Boltzman Distribution")
plt.legend()
plt.show()
```
